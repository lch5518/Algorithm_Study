## 문제
패밀리 레스토랑에 가서, 여러 개의 테이블에 사람을 나누어 앉게 하려고 합니다. <br>
이때, 한 사람만 앉는 테이블이 없게 그룹을 지어야 합니다.<br>
인원수를 나누는 패턴만 구하면 되며, 누가 어디에 앉는지 등은 고려하지 않아도 괜찮습니다. <br>
예를 들어 6명이라면, 다음과 같은 4가지가 됩니다.<br>
 * 2명 + 2명 + 2명
 * 2명 + 4명
 * 3명 + 3명
 * 6명
 
 한 개의 테이블에 앉을 수 있는 사람은 최대 10명입니다.<br>
 100명이 하나 이상의 테이블에 나누어 앉는 패턴을 구하세요

```python
M = 10
N = 100
```

```python
def check(remain, pre):
    if remain < 0:
        return 0
    elif remain == 0:
        return 1
    
    # 재귀적 처리
    cnt = 0
    for i in range(pre, M + 1):
        cnt += check(remain - i, i)
    return cnt
```

```python
print(check(6,2),'가지')
```

```python
print(check(N,2),'가지')
```

사람 수가 많아지면 처리 시간이 매우 오래 걸리기 때문에 한 번 계산한 결과를 메모해두고, 필요할 때 새로 계산하지 않고 사용해봅시다.<br>


```python
memo = {}
def check(remain, pre):
    # 이전에 계산한 적이 있다면, 메모했던 값을 반환
    key = str([remain, pre])
    if key in memo:
        return memo[key]
    
    # 배치할 사람이 더 이상 없으면 종료
    if remain < 0:
        return 0
    elif remain == 0:
        return 1
    
    # 재귀적 처리
    cnt = 0
    for i in range(pre, M + 1):
        cnt += check(remain - i, i)
    # 계산 결과를 메모하면서 반환
    memo[key] = cnt
    return cnt
```

```python
print(check(N,2),'가지')
```

이중 반복문 사용

```python
M = 10
N = 100
table = [0] * (M + 1)
```

```python
table
```

```python
for i in range(0, M + 1):
    table[i] = [0] * (N + 1)
    table[i][0] = 1
```

```python
import pandas as pd
```

```python
tableView = pd.DataFrame(table)
tableView
```

```python
for i in range(1, M + 1):
    for j in range(2, N + 1):
        if((i >= 2) and (j >= i)):
            table[i][j] = table[i][j - i]
        if i > 2:
            table[i][j] += table[i - 1][j]
```
