```python
import pandas as pd


class productCount:
    def __init__(self):        
        self.df = pd.DataFrame(data=[],columns=['제품명','수량'])
    def print_menu(self):
        print('제품수량관리\n')
        print('============================================================\n')
        print('1. 입력 2. 출력 3.검색  9.종료 중에 메뉴를 선택하세요')
        print('============================================================\n')
        menu = int(input('입력하세요  -> '))
        return int(menu)
    
    def set_input(self):
        nextItem = True
        while(nextItem==True):
            productName = str(input('제품명 : '))
            productCount = int(input('수량 : '))
            productSeries = pd.Series([productName,productCount],index=['제품명','수량'])
            self.df = self.df.append(productSeries,ignore_index=True)
            continueItem = str(input('계속 입력? Y/N'))
            if(continueItem=='Y'or continueItem=='y'):
                continue
            elif(continueItem=='N'):
                break
            else:
                break
                
    def set_print(self):
        print(self.df)
    
    def select_product(self):
        select = str(input('제품명 : '))
        df_boolean = self.df[self.df['제품명']==select]
        print(df_boolean)
    def Function(self):
        while 1:
            print('\n\n')
            menu = self.print_menu()
            if menu == 1:
                self.set_input()
            elif menu == 2:
                self.set_print()
            elif menu == 3:
                self.select_product()
            elif menu == 9:
                break
                
if __name__ == '__main__':
    productcount = productCount()
    productcount.Function()
```
